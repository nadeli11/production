```python

import os
import glob
from dotenv import load_dotenv
import dask.dataframe as dd

# Load environment variables
load_dotenv()

# Get the path from .env
price_data_dir = os.path.abspath(os.getenv("PRICE_DATA"))

# Check the directory
if not os.path.isdir(price_data_dir):
    raise FileNotFoundError(f"Directory does not exist: {price_data_dir}")

# Recursively find all .parquet files inside subfolders
parquet_files = glob.glob(os.path.join(price_data_dir, "**", "*.parquet"), recursive=True)

# Debug print
print(f"Found {len(parquet_files)} .parquet file(s).")

if not parquet_files:
    raise FileNotFoundError(f"No .parquet files found recursively under: {price_data_dir}")

# Load into Dask
df = dd.read_parquet(parquet_files)

# Preview
print(df.head())
```

For each ticker and using Dask, do the following:

+ Add lags for variables Close and Adj_Close.
+ Add returns based on Close:
    
    - `returns`: (Close / Close_lag_1) - 1

+ Add the following range: 

    - `hi_lo_range`: this is the day's High minus Low.

+ Assign the result to `dd_feat`.

(4 pt)

```python
# Write your code below.

# Reading .parquet files with selected columns
columns_to_use = ["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume", "ticker"]

try:
    ddf = dd.read_parquet(parquet_files, engine="pyarrow", columns=columns_to_use)
except Exception as e:
    raise RuntimeError(f"Failed to read parquet files: {e}")

# Renaming columns for consistency
ddf = ddf.rename(columns={"Adj Close": "Adj_Close"})

# Filtering to a few tickers to avoid memory overload (adjust as needed)
tickers_to_keep = ["AAPL", "MSFT", "GOOG"]
ddf = ddf[ddf["ticker"].isin(tickers_to_keep)]

# Ensuring date is datetime and sort
ddf["Date"] = dd.to_datetime(ddf["Date"])
ddf = ddf.sort_values(by=["ticker", "Date"])

# Creating lag features
ddf["Close_lag_1"] = ddf.groupby("ticker")["Close"].shift(1)
ddf["Adj_Close_lag_1"] = ddf.groupby("ticker")["Adj_Close"].shift(1)

# Calculating returns and hi_lo_range
ddf["returns"] = (ddf["Close"] / ddf["Close_lag_1"]) - 1
ddf["hi_lo_range"] = ddf["High"] - ddf["Low"]

# Repartition after filtering (safe memory)
ddf = ddf.repartition(npartitions=10).persist()
```

+ Convert the Dask data frame to a pandas data frame. 
+ Add a new feature containing the moving average of `returns` using a window of 10 days. There are several ways to solve this task, a simple one uses `.rolling(10).mean()`.

(3 pt)

```python
# Write your code below.

# Computing to pandas
try:
    pdf = ddf.compute()
except MemoryError:
    raise MemoryError("Dataset is too large to fit in memory — filter further.")

# Final sort before rolling
pdf = pdf.sort_values(by=["ticker", "Date"])

# Calculating 10-day rolling average of returns
pdf["returns_ma_10"] = pdf.groupby("ticker")["returns"].transform(lambda x: x.rolling(window=10).mean())

# Reviewing results
print("Pipeline completed successfully. Here's a preview:")
print(pdf.head())
```
